```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer #text to numbers
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data colection and pre processing

```python
raw_mail_data= pd.read_csv('/content/mail_data.csv')
```

```python
#replace the null valules with null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.head()
```

label encoding

```python
#spam as 0 and ham as 1
mail_data.loc[mail_data['Category']== 'spam','Category', ] =0
mail_data.loc[mail_data['Category']== 'ham', 'Category',] = 1
```

```python
mail_data.tail()
```

```python
X= mail_data['Message']
Y= mail_data['Category']
```

```python
print(Y)
```

```python
Y= Y.astype('int') #object type to int for model fitting
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
print(X_train_features)
```

```python
model= LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

```python
#prediction on training data
predicted_training_values = model.predict(X_train_features)
training_accuracy = accuracy_score(Y_train,predicted_training_values)
```

```python
print(training_accuracy)
```

```python
#prediction on testing data
predicted_testing_values = model.predict(X_test_features)
training_accuracy = accuracy_score(Y_test,predicted_testing_values)
```

```python
print(training_accuracy)
```

Building a predictive system

```python
from scipy.spatial import transform
input_mail= ["Here's something most students don't know:The best internships are never posted publicly. They go to whoever reached out first — directly to the founder or hiring manager — before anyone else even knew the role existed."]

#converting text into features
input_mail_features = feature_extraction.transform(input_mail)

prediction = model.predict(input_mail_features)

print(prediction)

if(prediction==1):
    print('ham mail')

else:
    print('spam mail')
```
