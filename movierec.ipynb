```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import difflib
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_data= pd.read_csv('/content/movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
imp_features = ['director','genres','keywords','cast','tagline']
```

```python
for features in imp_features:
  movie_data[features]=  movie_data[features].fillna('')
```

```python
print(imp_features)
```

```python
combined_features = movie_data['genres']+  movie_data['keywords']+movie_data['cast']+movie_data['tagline']+movie_data['director']
```

```python
print(combined_features)
```

```python
vectoriser = TfidfVectorizer()
feature_vectors= vectoriser.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape #every one vector comaprison with every other
```

Movie recommendation system

```python
#input movie from user
input_movie= input('Enter your favourite movie: ')

#make list of the movies in the data
list_of_all_movies = movie_data['title'].tolist()

#finding close match of the input movie
find_closest_match = difflib.get_close_matches(input_movie, list_of_all_movies)
closest_match = find_closest_match[0]
print('Oh, you meant this: ', closest_match)

#finding index of the closest match
index_of_the_movie = movie_data[movie_data.title == closest_match]['index'].values[0]

#score of movies wrt to the index_of_the_movie
score_of_similar_movies= list(enumerate(similarity[index_of_the_movie]))

#descending order wrt score_of_similar_movies[1]
sorted_list = sorted(score_of_similar_movies, key= lambda x:x[1], reverse = True)

#printing the similar movies
print('Here are few recommended movies for you: ')
i=1
for movies in sorted_list:
  index= movies[0]
  name_of_the_movie = movie_data[movie_data.index == index]['title'].values[0]
  if(i<=30):
    print(i, name_of_the_movie)
    i+=1




```
